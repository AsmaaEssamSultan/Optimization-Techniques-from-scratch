```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

#### Generate data

```python
X = np.linspace(0,20,30).reshape(30,1)
Y = (-1*X+2).reshape(30,1)
```

#### Plot data points

```python
plt.scatter(X,Y)
plt.grid()
plt.show()
```

## RMSProp

```python
def RMSProp(X, Y, alpha, Beta, Epsilon, Iterations):
    theta_0 , theta_1 = 0 , 0
    Vt0 = 0
    Vt1 = 0
    m = X.size
    J_L = []
    H_L = []
    theta_0L = [0]
    theta_1L = [0]
    for i in range(Iterations):
        h = theta_0 + theta_1 * X
        H_L.append(h)
    
        J = (1/(2*m))*sum((h - Y)**2)
        J_L.extend(J)

        dtheta_0 = (1/m)*sum(h - Y)
        dtheta_1 = (1/m)*sum((h - Y)* X)
        
        Vt0 = Beta*Vt0 + (1-Beta)*dtheta_0**2
        Vt1 = Beta*Vt1 + (1-Beta)*dtheta_1**2
        
        dtheta = np.array([dtheta_0,
                          dtheta_1])
        if np.linalg.norm(dtheta) < 0.1:
            break
        
        if i > 1 and abs(J_L[-2]- J_L[-1]) < 0.001:
            break
    
        if i < Iterations-1:
            theta_0 = theta_0 - (alpha/(np.sqrt(Vt0)+Epsilon))*dtheta_0
            theta_0L.extend(theta_0)
            theta_1 = theta_1 - (alpha/(np.sqrt(Vt1)+Epsilon))*dtheta_1
            theta_1L.extend(theta_1)
            
    return theta_0L, theta_1L, J_L, H_L, theta_0, theta_1
```

```python
alpha = 0.002
Epsilon = 1e-8
Beta = 0.8
Iterations = 100000
theta_0L, theta_1L, J_L, H_L, theta_0, theta_1 = RMSProp(X, Y, alpha, Beta, Epsilon, Iterations)
```

```python
print(theta_0)
print(theta_1)
```

```python
r2_score(Y,theta_0+ theta_1*X)
```
